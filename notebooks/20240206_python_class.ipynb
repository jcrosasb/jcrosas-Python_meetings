* 'function2' does work, because the '+' method is defined for strings

CONCLUSION:  Neither function nor function2 will care if the argument is an Int or not because Python is a dynamically typed lenguage. ## Primitive and reference values
* Primitive value: a value without any methods of any kind. Just the actual value
* Reference value: usually an object of some kind, an instance of a class

Python treats everything as an object, so no primitive values

## Noop Function
No operation function

````python
def noop():
    pass
````

EXAMPLE OF NOOP function use:

```
def process_data(data, callback=noop):
    # Process the data
    # Call the callback function (or noop if not provided)
    callback()

# Call process_data with a custom callback function
process_data(my_data, my_callback)

# Call process_data without a callback function (uses noop)
process_data(my_data)
```

In the second call to process_data, if no callback function is provided, the noop function is used as the default value for the callback parameter, ensuring that the function always has a callable value to execute.

## Stub function
A function whose implementation is not there yet

EXAMPLE OF STUB FUNCTION:

```
def get_number_from_user():
    # Stub function for getting a number from the user
    # Returns a predetermined value for testing purposes
    return 5  # Stubbed value for testing

def calculate_square():
    # Function that calculates the square of a number obtained from the user
    number = get_number_from_user()  # Call the stubbed function
    return number ** 2

# Test the calculate_square function
result = calculate_square()
print("Square of the number:", result)
```

Function `get_number_from_user` is stubbed, the `calculate_square` function can be tested independently of user input.
